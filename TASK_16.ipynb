```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv("/content/factory_worker_safety_multimodal_dataset_v2.csv")
```

```python
df = df.drop(['Report_ID','Image_Path','Video_Path','Audio_Path','Incident_Description'], axis=1)
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for col in df.columns:
  if df[col].dtype=='object':
    df[col]=le.fit_transform(df[col])
```

```python
x=df.drop("Risk_Level",axis=1)
```

```python
y=df["Risk_Level"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
m=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
m.fit(x_train,y_train)
```

```python
yp=m.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
a=accuracy_score(y_test,yp)
a
```

```python
classification_report(y_test,yp)
```

```python
import joblib
joblib.dump(m,"Workersaftey.pkl")
```
