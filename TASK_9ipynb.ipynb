LOADING AND IMPORTING THE DATASET

```python
from sklearn import datasets
iris = datasets.load_iris()
```

SEE THE DIRECTORY

```python
dir(iris)
```

SEEING THE FIRST COLUMN

```python
iris.data[0]
```

CREATE A MODEL

```python
from sklearn.linear_model import LogisticRegression
m=LogisticRegression()
```

IMPLEMENT THE TRAINA AND TEST

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2)
```

HOW MUCHH DATA TO BE TAKEN AS TRAIN

```python
len(x_train)
```

HOW MUCH DATA TOBE TAKEN AS TEST

```python
len(x_test)
```

TRAIN THE MODEL

```python
m.fit(x_train,y_train)
```

CHECKING THE SCORE

```python
m.score(x_test,y_test)
```

VIEWING THE MODEL PREDICTED ONE

```python
m.predict(iris.data[0:5])
```

MODEL PREDICTED OUTPUT

```python
y=m.predict(x_test)
y
```

IMPLEMENTING THE CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y)
cm
```

VISUALIZING THE PREDICTED

```python
import seaborn as sn
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
