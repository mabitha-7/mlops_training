```python
import pandas as pd
```

```python
df=pd.read_csv("/content/factory_worker_safety_multimodal_dataset_v2.csv")
df.head()
```

```python
df.info()
```

```python
x= df[['PPE_Compliance_Score','Audio_dB_Level','Audio_Confidence','Video_Confidence','Text_Word_Count']]
```

```python
y = df['Fusion_Risk_Score']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
m=LinearRegression()
```

```python
m.fit(x_train,y_train)
```

```python
yp=m.predict(x_test)
```

```python
yp[:5]
```

```python
y_test[:5]
```

```python
s=m.score(x_test,y_test)
s
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, yp)
mse
```
